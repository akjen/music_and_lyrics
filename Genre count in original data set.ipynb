```python
import pandas as pd
import csv
import os
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from operator import itemgetter

os.chdir(r'C:\Users\Gard\Documents\Bygg- og miljøteknikk\Høst 2020\Maskinlæring\Project')

data = csv.reader(open('combined_data_with_semicolon.csv'), delimiter=',')
next(data)
genre_count = [['Total words',0]]

for row in data:
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    words = tokenizer.tokenize(row[3])
    #print(words)
    for word in words:
        if word in np.asarray(genre_count,dtype=object)[:,0]:
            genre_count = np.asarray(genre_count, dtype=object)
            genre_count[np.where(genre_count==word),1] += 1
            genre_count = genre_count.tolist()
        else:
            new_entry = [word, 1]
            genre_count.append(new_entry)

genre_count = np.asarray(sorted(genre_count, key=itemgetter(1), reverse=True), dtype=object)       

print(genre_count)
```
